```python
# Fungsi untuk mengidentifikasi kata-kata tidak dikenal
def identify_unknown_words(text, known_words):
    # Tokenisasi kata-kata dalam teks
    words = text.split()

    # Identifikasi kata-kata tidak dikenal
    unknown_words = [word for word in words if word.lower() not in known_words]

    return unknown_words

# Contoh kamus kata-kata yang dikenal (dapat disesuaikan dengan kebutuhan)
known_words = ['yang', 'banyak', 'nyaman', 'pokoknya', 'oke', 'banget']

# Contoh teks dengan kata-kata aneh
example_text = "Tempat ini U bnyk yg nyaman, pokoknya X banget!"

# Identifikasi kata-kata tidak dikenal
unknown_words = identify_unknown_words(example_text, known_words)

# Output hasil identifikasi
print("Teks Asli                :", example_text)
print("Kata-kata Tidak Dikenal:", unknown_words)
```

```python
import subprocess
import json

def cari_kata_kunci(kata_kunci):
    try:
        # Eksekusi skrip JavaScript menggunakan kbbi.js
        result = subprocess.check_output(['node', '-e', f'const KBBI = require("kbbi.js"); KBBI.cari("{kata_kunci}").then(console.log).catch(console.error);'], text=True)

        # Parsing hasil JSON
        result_json = json.loads(result)
        return result_json
    except Exception as e:
        return {'error': str(e)}

# Contoh penggunaan
kata_kunci = "pohon"
hasil_pencarian = cari_kata_kunci(kata_kunci)
print(hasil_pencarian)
```
